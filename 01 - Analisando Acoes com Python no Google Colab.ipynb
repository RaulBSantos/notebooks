# Obtendo os dados do mercado

```
ibov = web.get_data_yahoo('^BVSP')
```

```
ibov.head()
```

```
ibov.tail()
```

```
ibov['Close'].plot(figsize=(24,8), label='IBOV')
ibov['Close'].rolling(21).mean().plot(label='Média 21d')
ibov['Close'].rolling(200).mean().plot(label='Média 200d')
plt.legend()
```

```
ibov_fatiado = ibov[ibov.index.year == 2008]
ibov_fatiado['Close'].plot(figsize=(24,8), label='IBOV')
ibov_fatiado['Close'].rolling(21).mean().plot(label='Média 21d')
ibov_fatiado['Close'].rolling(200).mean().plot(label='Média 200d')
plt.legend()
```

```
ibov_fatiado = ibov[(ibov.index.year >= 2008) & (ibov.index.year <= 2015)]
ibov_fatiado['Close'].plot(figsize=(24,8), label='IBOV')
ibov_fatiado['Close'].rolling(21).mean().plot(label='Média 21d')
ibov_fatiado['Close'].rolling(200).mean().plot(label='Média 200d')
plt.legend()
```
